```python
import pandas as pd
```

```python
df = pd.read_csv('../data_final_to_preprocess.csv')
```

Clean 'comma' srings to float. These are the columns:
final_exam
ga1
ga2
homework
And remove rows with 'ONG'

```python
cols = ['final_exam', 'ga1', 'ga2', 'homework']


# Remove rows where any of the selected columns contain 'ONG'
df = df[~df[cols].isin(['ONG']).any(axis=1)]

# Convert valid numeric strings to floats
for col in cols:
    df[col] = df[col].astype(str).str.replace(',', '.').astype(float)
```

```python
# Create boolean column based on nr_practice_exams_viewed
df['bool_practice_exams_viewed'] = df['nr_practice_exams_viewed'].astype(float) > 0
```
